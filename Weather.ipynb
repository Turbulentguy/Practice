2. Can you plot the trend of temperature (Temp_C) over time?

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

plt.plot(df['Date/Time'], df['Temp_C'])

plt.xticks(rotation = 90)

plt.tight_layout()

plt.grid(True)

plt.show()
```

3. What is the distribution of relative humidity (Rel Hum_%)? Plot a histogram to visualize it.

```python
plt.hist(df['Rel Hum_%'])

plt.xlabel('Percentage of Relative Humidity')

plt.ylabel('Frequency')

plt.grid(True)

plt.show()
```

4. Create a box plot to show the distribution of wind speed (Wind Speed_km/h). Are there any outliers?

```python
plt.boxplot(df['Wind Speed_km/h'])

plt.grid(True)

plt.show()
```

```python
import numpy as np

Q1 = df['Wind Speed_km/h'].quantile(0.25)

Q3 = df['Wind Speed_km/h'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR

upper_bound = Q3 + 1.5 * IQR

df[(df['Wind Speed_km/h'] < lower_bound) | (df['Wind Speed_km/h'] > upper_bound)]
```

7. What are the frequencies of different weather conditions? Visualize this with a bar chart.

```python
df.head()
```

```python
plt.bar(df['Weather'].value_counts().index , df['Weather'].value_counts().values)

plt.xticks(rotation = 90)

plt.xlabel('Weather')

plt.ylabel('Frequency')

plt.show()
```

8. Plot a line graph showing the trend of atmospheric pressure (Press_kPa) over time.

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

plt.plot(df['Date/Time'] , df['Press_kPa'])

plt.xlabel('Date & Time')

plt.ylabel('Press (kPa)')

plt.grid(True)

plt.show()
```

9. How does visibility (Visibility_km) vary with different weather conditions? Use a grouped bar chart or box plot.

```python
grouped = df.groupby(['Visibility_km','Weather']).size().unstack(fill_value = 0)

grouped.plot(kind = 'bar', stacked = True)
```
